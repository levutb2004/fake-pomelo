```python
import os
from osgeo import gdal
import numpy as np
dir_path = r"E:\fake-pomelo\pomelo_input_data\VNM_commune\covariates"
files = os.listdir(dir_path)

for file in files:
    ds = gdal.Open(os.path.join(dir_path,file), gdal.GA_ReadOnly)
    band = ds.GetRasterBand(1)
    nodata = band.GetNoDataValue()
    arr = band.ReadAsArray().astype(np.float64)
    mask = arr != nodata
    valid = arr[mask]

    mean = float(valid.mean())
    std = float(valid.std())
    print(file, mean, std)
```
